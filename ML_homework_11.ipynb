Transponowana konwolucja to operacja pozwalające na zwiększenie rozdzielczości mapy cech (upsampling) przy użyciu jądra - jest to w pewnym sensie odwrócenie procesu standardowej konwolucji. Tu jednak, zamiast przemieszczania jądra po inpucie i obliczania ioczynów i sum, postępujemy następująco:



*(Uwaga: Dla uproszczenia zakładam, że input i jądro są kwadratami, padding jest taki sam ze wszystkich stron i stride jest taki sam w kierunku pionowym i poziomym.)*



1.   Nasz input $A$ ma wymiary $n \times n$, jądro $K$ ma wymiary $k \times k,$ padding jest równy $p,$ a stride jest równy $s.$ Stride mówi o ile przesuwamy nasze jądro przy obliczaniu outputu, padding mówi jak grubą ramkę z dodatkowych wierszy/kolumn wpisujemy wokół outputu podczas obliczeń, kernel size to po prostu rozmiar jądra.
2.   Niech $m=(n-1)s-2p+k.$ Tworzymy macierz $m \times m$ wypełnioną zerami - w następnych krokach będziemy ją wypełniać liczbami.
3. Dla każdego wyrazu $a_{ij}$ macierzy wejściowej tworzymy macierz $a_{ij} \cdot K$ i umieszczamy ją na naszej zerowej macierzy $m \times m$ tak, by lewy górny róg znajdował się na pozycji $((i-1)s+1,(j-1)s+1).$
4. W ten sposób uzyskujemy $n^2$ macierzy $m \times m$ - dodajemy je wszystkie i uzyskujemy pojedynczą macierz $ m \times m.$
5. Z góry, dołu, lewej i prawej usuwamy po $p$  wierszy/kolumn i dostajemy w wyniku macierz wyjściową.
