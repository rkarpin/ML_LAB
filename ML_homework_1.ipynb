```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/SzymonNowakowski/Machine-Learning-2024/refs/heads/master/PCA_three_clusters_data.csv')
```

```python
data
```

```python
x = data["x"]
y = data["y"]
z = data["z"]
classes = data["class"]
```

```python
plt.scatter(x,y,c=classes,cmap='brg', alpha =0.7)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
plt.scatter(y,z,c=classes,cmap='brg', alpha =0.7)
plt.xlabel("y")
plt.ylabel("z")
plt.grid(True)
plt.show()
plt.scatter(x,z,c=classes,cmap='brg', alpha =0.7)
plt.xlabel("x")
plt.ylabel("z")
plt.grid(True)
plt.show()
```

```python
df = data[['x','y','z']]
df=(df-df.mean())/df.std()
pca = PCA(n_components = 3)
df = pca.fit_transform(df)
df=df.T
```

```python
plt.scatter(df[0],df[1],c=classes,cmap='brg', alpha =0.7)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.grid(True)
plt.show()
```
